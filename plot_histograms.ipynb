```python
import numpy as np
import matplotlib.pyplot as plt

SMALL_SIZE = 16
MEDIUM_SIZE = 20
BIGGER_SIZE = 24

plt.rc('font', size=SMALL_SIZE)          # controls default text sizes
plt.rc('axes', titlesize=SMALL_SIZE)     # fontsize of the axes title
plt.rc('axes', labelsize=MEDIUM_SIZE)    # fontsize of the x and y labels
plt.rc('xtick', labelsize=SMALL_SIZE)    # fontsize of the tick labels
plt.rc('ytick', labelsize=SMALL_SIZE)    # fontsize of the tick labels
plt.rc('legend', fontsize=SMALL_SIZE)    # legend fontsize
plt.rc('figure', titlesize=BIGGER_SIZE)  # fontsize of the figure title
```

```python
d = np.load('histogram_input.npz')
```

```python
fig = plt.Figure(figsize=(12,6), facecolor='white')

xmax = 3.
nbins = 70

ax = fig.add_subplot(121)
bins = np.linspace(0., xmax, nbins)
ax.hist(d['nhits_pred']/d['nhits_truth'], bins=bins, density=True, label='All')
ax.hist((d['nhits_pred']/d['nhits_truth'])[d['is_strict_em_truth'] & (d['esum_truth']>20)], bins=bins, alpha=.5, density=True, label='EM only')
ax.set_xlabel('$N_{hits}^{pred}/N_{hits}^{tru}$')
ax.set_xlim(0., xmax)
ax.legend()

ax = fig.add_subplot(122)
bins = np.linspace(0., xmax, nbins)
ax.hist(d['esum_pred']/d['esum_truth'], bins=bins, density=True, label='All')
ax.hist((d['esum_pred']/d['esum_truth'])[d['is_strict_em_truth'] & (d['esum_truth']>20)], bins=bins, alpha=.5, density=True, label='EM only')
ax.set_xlabel('$E_{hits}^{pred}/E_{hits}^{tru}$')
ax.set_xlim(0., xmax)
ax.legend()

fig
```
